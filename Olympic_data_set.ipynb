Athlete age analysis

```python
from datetime import datetime
```

```python
df['birth_date'].dtype
```

```python
# converting birth date (string >> birthdat (datetime))
```

```python
df['birth_date'] = pd.to_datetime(df['birth_date'])
```

```python
df['birth_date'].dtype
```

```python
# creating a lambda
```

```python
datetime.now()
```

```python
df['birth_date'] = pd.to_datetime(df['birth_date'])

df['age'] = df['birth_date'].apply ( lambda birth_date : ( datetime.now() - birth_date ) .days // 365)
```

```python
df[['name','age']].head()
```

Disply the name of the youngest/oldest athleste

***idxmin and idxmax***

```python
df['age']
```

```python
df['name']
```

.loc[]  >> location of data

.iloc[]  >> numbers *italicized text*

```python
youngest = df.loc[df['age'].idxmin()]
oldest = df.loc[df['age'].idxmax()]
print(f'Youngest person in the Olympics is {youngest['name']} and his age is {youngest['age']} years old.')
print(f'The oldest person is {oldest['name']} and he is {oldest['age']} years old.')
```
