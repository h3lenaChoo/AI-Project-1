# 1. Rugplot -> 1 continuous variable

* find average

```python
# rugplot puts down 1 mark/point per data

# 1 line represents 1 person

sns.rugplot(x= 'salary', data = df)

# density higher -> determine the average (salary
```

```python
sns.rugplot(x='division',data=df)

# if u apply rugplot to a non-continuous variable= no use/meaningless
```

# 2. Displot() and Histplot()

*Histplot()*
* make it obvious
* average in middle

*Distplot*

```python
sns.histplot(x='salary',data=df)
```

```python
sns.displot(x='salary',data=df)
```

```python
sns.histplot(data=df, x ='salary', kde = True)

# kde -> kernal density estimator (where in the graph is data most concentrated)

mean_value = np.mean(df['salary'])

# plt. axvline (vertical)  * top to bottom
# plt.  axhline (horizontal)

plt.axvline(mean_value, color = 'red', ls = '--')

# text(x,y,text)
plt.text(mean_value * 1.05, 100, 'Mean Value')

plt.title("Histogram ....")
```

```python
# application -> show where exactly is the MEAN

# i) store min value --> np.mean()
```

```python
# the smaller amount of BINS the larger the rannge per BINS
```

```python
# size of bins is crucial to show info/ concentration/range of norms/ AVERAGE
```

```python
sns.histplot(data=df,x='salary',bins=10  )
```

adding grid -> sns.set()

```python
sns.set(style = 'darkgrid')

sns.displot(data=df, x ='salary')
```

```python
sns.displot(data=df, x = 'salary', bins = 20, color = 'red', lw='6', edgecolor = 'black', alpha = 0.5, ls = '--')
```
