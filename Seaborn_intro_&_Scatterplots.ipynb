```python
#use matplotlib in our scatterplot
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= 'sales', y='salary', data=df, hue = 'level of education') # hue >> colors
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= 'sales', y='salary', data=df, hue = 'work experience') # hue >> colors
```

```python
# change color mapping >> # https://matplotlib.org/stable/users/explain/colors/colormaps.html
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= 'sales', y='salary', data=df, hue = 'work experience', palette = 'plasma', s = 200)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= 'sales', y='salary', data=df, hue = 'work experience', palette = 'plasma', s = 200,alpha = 0.2 )
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x= 'sales', y='salary', data=df, palette = 'plasma', s = 150, style = 'level of education', hue = 'level of education')
```
