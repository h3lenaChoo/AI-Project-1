categorical data - data in distrinct grouping (non-relevant)

```python
#categorical plots will display 2 things

# i) mean
# ii) count

# countplot () - count the number of rows per category   --> non-continuous
# barplot () - general form of displaying any data (mean or std)
```

```python
df['division'].unique()
```

```python
plt.figure(figsize=(12,6))

sns.countplot( x = 'division', data = df)

# y axis will become the count
# x axis will be the categorical data
```

```python
plt.figure(figsize=(12,6))

sns.countplot( x = 'level of education', data = df)
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(12,6))

sns.countplot( x = 'level of education', data = df, hue = 'division')
```
