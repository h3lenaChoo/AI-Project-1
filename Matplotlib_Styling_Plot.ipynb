```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = x * 2
```

legends

```python
# labeling
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1]) #position >> more than 1 axes

axes.plot(x,x ** 2, label = 'x^2')
axes.plot(x,x ** 3,label = 'x^3')

axes.legend()
```

```python
axes.legend(loc = 1) # top right
axes.legend(loc =2) #top left
axes.legend(loc = 0)

fig
```

setting colors, linewidths, linetypes

colors with MatLab like syntax

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-') #blue line with dots
ax.plot(x, x**3, 'b.-')
```

```python
fig, ax = plt.subplots()

ax.plot(x,x ** 2, color = 'blue') # blue line with dots
ax.plot(x,x**3,color = '#CC0066')
```

width/ thickness of line

line width >> lw

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x,x-1, color = 'red', lw = 10, alpha = 0.2) # more lower the alpha, the more transparent the line will be
ax.plot(x,x-1, color = 'red', lw = 5)
ax.plot(x,x-1, color = 'red', lw = 1)
```

linestyles

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x,x-1, color = 'red', lw = 10, alpha = 0.2, ls = '-.') # more lower the alpha, the more transparent the line will be
ax.plot(x,x-1, color = 'red', lw = 5, ls = ':')
ax.plot(x,x-1, color = 'red', lw = 1, ls = '-.')
```

markers

```python
# https://matplotlib.org/3.2.2/api/markers_api.html
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x,x-1, color = 'red', lw = 5, alpha = 0.2, marker = "+") # more lower the alpha, the more transparent the line will be
ax.plot(x,x-2, color = 'green', lw = 5, ls = ':', marker = 'o')
ax.plot(x,x-3, color = 'blue', lw = 5, ls = '-.', markersize = 20)
```
